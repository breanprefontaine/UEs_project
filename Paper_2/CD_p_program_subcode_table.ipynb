# CD p/program and subcode table



### Imports

```python
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker
```

### Data 


### CD p/program
CD = [%Q2Q, %PISEC, %ST]

```python
"""-------------Data for graph----------------"""
AE = [37,53,39]
ME = [17,25,19]
NR = [46,22,42]

# Set bottom of NR for stacking
NR_bottom = [i+j for i,j in zip(AE,ME)]


"""-------------Data for table----------------"""
#                            Q2Q                             PISEC                             ST
#                  AE         ME         NR         AE         ME        NR          AE        ME        NR
cell_text = [['', 55.5,'','',20.9,'','',38.8,'','',13.1,'','', 5.0,'','', 0.0,'','',14.0,'','', 1.4,'','',13.0,''],  #Insider
             ['', 28.2,'','',72.1,'','',16.5,'','',57.1,'','',60.0,'','',51.4,'','',53.0,'','',54.8,'','',26.0,''],  #Inbound
             ['',  9.1,'','', 7.0,'','',27.3,'','', 7.1,'','',20.0,'','',22.9,'','',27.0,'','',39.7,'','',49.0,''],  #Neutral
             ['',  6.4,'','', 0.0,'','', 5.8,'','',17.9,'','', 0.0,'','', 8.6,'','', 2.0,'','', 1.4,'','', 0.6,''],  #Peripheral
             ['',  0.9,'','', 0.0,'','',11.5,'','', 4.8,'','',15.0,'','',17.1,'','', 4.0,'','', 2.7,'','',11.0,'']]  #Outbound

# Colours for table
cd_colours = ['sandybrown','sandybrown','sandybrown',
              'lawngreen','lawngreen','lawngreen',
              'mediumturquoise', 'mediumturquoise','mediumturquoise',
              'sandybrown','sandybrown','sandybrown',
              'lawngreen','lawngreen','lawngreen',
              'mediumturquoise', 'mediumturquoise','mediumturquoise',
              'sandybrown','sandybrown','sandybrown',
              'lawngreen','lawngreen','lawngreen',
              'mediumturquoise', 'mediumturquoise','mediumturquoise']

insider = ['darkorange','darkgreen','steelblue']

inbound = ['orange','green','dodgerblue']

neutral = ['sandybrown', 'lime', 'deepskyblue']

peripheral = ['gold','greenyellow','lightskyblue']

outbound = ['navajowhite','lightgreen','paleturquoise']

AE_locations = [0,1,2,9,10,11,18,19,20]
```

### Plotting


```python
# Set barwidth 
barWidth = 1.3

# Set tick positions for x and y axes, respectively
cd_positions = [1,2.89,4.83]
percentage_positions = [0,20,40,60,80,100]

# Tick labels for x and y axes, respectively
cd_labels = ['','AE','','','ME','','','NR','',
             '','AE','','','ME','','','NR','',
             '','AE','','','ME','','','NR','']

percentage_labels = ['0 %', '20 %', '40 %', '60 %', '80 %', '100 %']

fig, ax = plt.subplots(1,1)

# AE bars for each community
plot_1 = ax.bar(cd_positions, AE, color='sandybrown', hatch='//', 
                width=barWidth,label='AE')

for i in range(len(AE)):
    
    ax.annotate('{}%'.format(AE[i]),xy=(cd_positions[i], AE[i]/2), 
                xytext=(0, 0), textcoords="offset points", 
                ha='center', va='bottom', size=12, bbox=dict(boxstyle="round", 
                                                            fc="sandybrown"))


## ME bars for each community
plot_2 = ax.bar(cd_positions, ME,bottom=AE, color='lawngreen',hatch='..',
                width=barWidth,label='ME')

for i in range(len(ME)):
    
    ax.annotate('{}%'.format(ME[i]),xy=(cd_positions[i], AE[i] + ME[i]/2), 
                xytext=(0, 0), textcoords="offset points", 
                ha='center', va='bottom', size=12,bbox=dict(boxstyle="round", 
                                                            fc="lawngreen"))

# NR bars for each community
plot_3 = ax.bar(cd_positions, NR,bottom=NR_bottom, color='mediumturquoise',
       width=barWidth,label='NR')

for i in range(len(NR)):
    
    ax.annotate('{}%'.format(NR[i]),xy=(cd_positions[i], 
                NR_bottom[i] + NR[i]/2), xytext=(0, 0), 
                textcoords="offset points", ha='center', va='bottom', size=12,
                                bbox=dict(boxstyle="round", 
                                          fc="mediumturquoise"))

# Set graph title 
ax.set_title('CD p/program',fontsize=16)

# Set customsized x-tick labels
plt.xticks([])
ax.set_xlim(0,5.8)

# Set customised y-tick labels
ax.tick_params(axis='y', which='both',length=0,labelsize=14)
ax.yaxis.set_major_locator(ticker.FixedLocator(percentage_positions))
ax.yaxis.set_major_formatter(ticker.FixedFormatter(percentage_labels))

# Add legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels,loc='upper left',bbox_to_anchor=(0, 1),
          ncol=1, prop={'size':14},markerscale=1.1)

# Set grid 
ax.grid(axis='both')
ax.set_axisbelow(True)

# Adjust location of graph
plt.subplots_adjust(left=0.2, bottom=0.4)

"""---------------------------------------- TABLE -------------------------------------------------------------------------------------"""

## Merged headers
COP_header = plt.table(cellText=[['']*3],
                      colLabels=['Q2Q', 'PISEC', 'ST'],
                       loc='bottom',
                       rowLoc='center',
                       colLoc='center',
                       bbox=[0, -0.22, 1, 0.18])
# Set fontsize
COP_header.auto_set_font_size(False)
COP_header.set_fontsize(13)

# CD headers
CD_header = plt.table(cellText=[['']*27],
                     colLabels=cd_labels,
                     loc='bottom',
                     rowLoc='center',
                     colLoc='center',

                     bbox=[0,-0.28,1,0.15],
                     colColours=cd_colours)
# Set fontsize
CD_header.auto_set_font_size(False)
CD_header.set_fontsize(13)

# Subcode headers
rows = [[''],[''],[''],[''],['']]
subcodes = ['Insider', 'Inbound', 'Neutral', 'Peripheral', 'Outbound']

subcode_header = plt.table(cellText=rows,
                    rowLabels=subcodes,
                     loc='bottom',
                     rowLoc='center',
                     colLoc='center',
                     fontsize=12,
                     bbox=[0,-0.7,1,0.5])

# Data
table = plt.table(cellText=cell_text,
                      loc='bottom',
                      cellLoc='center',
                      rowLoc='center',
                      colLoc='center',
                      bbox=[0,-0.7,1,0.5])
# Set fontsize
table.auto_set_font_size(False)
table.set_fontsize(10)


""" If you want to make the table without hatch and colours, then comment out sections [1] and [2] """
""" and uncomment section [0]:"""

""" Section [0] to uncomment """
#for i in range(2,27,3):
#    print(i)
#    for j in range(5):
#        edges = 'R'
#        table[(j,i)].visible_edges = edges 
""" End of section [0] """


""" Section [1] to comment out """
for i in range(len(AE_locations)):
    
    table[(0,AE_locations[i])].set_facecolor(insider[0])
    table[(1,AE_locations[i])].set_facecolor(inbound[0])
    table[(2,AE_locations[i])].set_facecolor(neutral[0])
    table[(3,AE_locations[i])].set_facecolor(peripheral[0])
    table[(4,AE_locations[i])].set_facecolor(outbound[0])
    
    table[(0,AE_locations[i]+3)].set_facecolor(insider[1])
    table[(1,AE_locations[i]+3)].set_facecolor(inbound[1])
    table[(2,AE_locations[i]+3)].set_facecolor(neutral[1])
    table[(3,AE_locations[i]+3)].set_facecolor(peripheral[1])
    table[(4,AE_locations[i]+3)].set_facecolor(outbound[1])
    
    table[(0,AE_locations[i]+6)].set_facecolor(insider[2])
    table[(1,AE_locations[i]+6)].set_facecolor(inbound[2])
    table[(2,AE_locations[i]+6)].set_facecolor(neutral[2])
    table[(3,AE_locations[i]+6)].set_facecolor(peripheral[2])
    table[(4,AE_locations[i]+6)].set_facecolor(outbound[2])

""" End of section [1] """
# Hatches for table
AE_hatch = []
ME_hatch = []

# Create hatch patterns with decreasing density
for j in reversed(range(5)):
    AE_hatch.append("/"*j)  
    ME_hatch.append("."*j)

AE_hatch_locations = [0,2,9,11,18,20]

""" Section [2] to comment out"""
# Apply hatch 
for m in range(len(AE_hatch_locations)):
    
    table[(0,AE_hatch_locations[m])].set_hatch(AE_hatch[0])
    table[(1,AE_hatch_locations[m])].set_hatch(AE_hatch[1])
    table[(2,AE_hatch_locations[m])].set_hatch(AE_hatch[2])
    table[(3,AE_hatch_locations[m])].set_hatch(AE_hatch[3])
    table[(4,AE_hatch_locations[m])].set_hatch(AE_hatch[4])
    
    table[(0,AE_hatch_locations[m]+3)].set_hatch(ME_hatch[0])
    table[(1,AE_hatch_locations[m]+3)].set_hatch(ME_hatch[1])
    table[(2,AE_hatch_locations[m]+3)].set_hatch(ME_hatch[2])
    table[(3,AE_hatch_locations[m]+3)].set_hatch(ME_hatch[3])
    table[(4,AE_hatch_locations[m]+3)].set_hatch(ME_hatch[4])
""" End of section [2] """


# Hatch for column titles
for p in range(len(AE_hatch_locations)):
    CD_header[(0,AE_hatch_locations[p])].set_hatch('//')
    CD_header[0,AE_hatch_locations[p]+3].set_hatch('..')

```
